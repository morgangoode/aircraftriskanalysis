Replot the previous barcharts and plot new ones to compare. 

```python
import matplotlib.pyplot as plt
```

Comparing data from all countries to USA only data.

First up, all countries:

```python
# set up plots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 5))

# create variables for easier reuse
all_countries_aircraft_type_counts = df_unique_aircraft_event_all_countries['aircraft category'].value_counts()
all_countries_event_type_counts = df_unique_aircraft_event_all_countries['ev_type'].value_counts()

# plot data
ax1.bar(all_countries_aircraft_type_counts.index, all_countries_aircraft_type_counts.values)
ax2.bar(all_countries_event_type_counts.index, all_countries_event_type_counts.values)

# customize appearance
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=45)
ax1.set_ylabel("Count of Events (Accidents + Incidents)")
ax2.set_ylabel("Count of Events")
ax1.set_title("Distribution of Event by Aircraft Type (All Countries)")
ax2.set_title("Accidents vs Incidents (All Countries)");
```

And now, USA only:

```python
# set up plots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 5))

# create variables for easier reuse
USA_aircraft_type_counts = df_unique_aircraft_events_usa['aircraft category'].value_counts()
USA_event_type_counts = df_unique_aircraft_events_usa['ev_type'].value_counts()

# plot data
ax1.bar(USA_aircraft_type_counts.index, USA_aircraft_type_counts.values)
ax2.bar(USA_event_type_counts.index, USA_event_type_counts.values)

# customize appearance
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=45)
ax1.set_ylabel("Count of Events (Accidents + Incidents)")
ax2.set_ylabel("Count of Events")
ax1.set_title("Distribution of Event by Aircraft Type (USA only)")
ax2.set_title("Accidents vs Incidents (USA only)");
```

There does not appear to be a significant difference. Now, let's do the same with `'flight purpose'`.

First up, all countries:

```python
# set up plots
fig, ax = plt.subplots(figsize=(40,10))


# create variable for easier reuse
all_countries_flight_purpose_counts = df_unique_aircraft_event_all_countries['flight purpose'].value_counts()


# plot data
ax.bar(all_countries_flight_purpose_counts.index, all_countries_flight_purpose_counts.values)


# customize appearance

ax.set_ylabel("Count of Events (Accidents + Incidents)")

ax.set_title("Distribution of Event by Flight Purpose (All Countries)");
```

Next up, USA only:

```python
# set up plots
fig, ax = plt.subplots(figsize=(40,10))


# create variable for easier reuse
USA_flight_purpose_counts = df_unique_aircraft_events_usa['flight purpose'].value_counts()


# plot data
ax.bar(USA_flight_purpose_counts.index, USA_flight_purpose_counts.values)


# customize appearance

ax.set_ylabel("Count of Events (Accidents + Incidents)")

ax.set_title("Distribution of Event by Flight Purpose (USA Only)");
```

Once again, no significant difference. This could be due to an oversampling of USA data. 

Let's look at the USA-only state-by-state breakdown:

```python
# set up plots
fig, ax = plt.subplots(figsize=(40,10))


# create variable for easier reuse
event_state_counts = df_unique_aircraft_events_usa['ev_state'].value_counts()


# plot data
ax.bar(event_state_counts.index, event_state_counts.values)


# customize appearance

ax.set_ylabel("Count of Events (Accidents + Incidents)")

ax.set_title("Distribution of Events by State");
```

The same states are in the top 8, the top ten shifts. Florida and Texas become equals whereas Texas was previously leading slightly. Also, the gap between Texas and Alaska has tightened.

This data begins in 1997, the year that launched the "Queen of the Skies" otherwise known as the Boeing 747. Boeing ending production for the 747 earlier this year citing the development of more fuel efficient two-engine widebody planes as the reason for discontinuing the formerly popular model.

What happens if we filter the data further? Perhaps limiting it to the last 5 years?

```python
df_unique_aircraft_events_usa.columns
```

```python
# filter by date to create a last 5 years df,
# starting in May 2018 until May 2023
df_unique_aircraft_events_usa_last_5 = df_unique_aircraft_events_usa[(df_unique_aircraft_events_usa['ev_date'] > "2018-04-30") & (df_unique_aircraft_events_usa['ev_date'] < "2023-06-01") ]
```

```python
# let's preview the work!
df_unique_aircraft_events_usa_last_5.head(3)
```

```python
df_unique_aircraft_events_usa_last_5.tail(3)
```

Now, let's plot again with our last 5 years (in the USA) dataframe. 

First up, a state-by-state breakdown of aircraft events.

```python
# set up plots
fig, ax = plt.subplots(figsize=(40,10))


# create variable for easier reuse
last_5_usa_event_state_counts = df_unique_aircraft_events_usa_last_5['ev_state'].value_counts()


# plot data
ax.bar(last_5_usa_event_state_counts.index, last_5_usa_event_state_counts.values)


# customize appearance

ax.set_ylabel("Count of Events (Accidents + Incidents)")

ax.set_title("Distribution of Events Involving Aircraft by State, May 1, 2018 through April 30, 2023");
```

Texas takes the lead! But we have the same states in the top 5.

Let's take a look at flight purpose in the USA over the last 5 years.

```python

# set up plots
fig, ax = plt.subplots(figsize=(40,10))


# create variable for easier reuse
last_5_usa_flight_purpose_counts = df_unique_aircraft_events_usa_last_5['flight purpose'].value_counts()


# plot data
ax.bar(last_5_usa_flight_purpose_counts.index, last_5_usa_flight_purpose_counts.values)


# customize appearance

ax.set_ylabel("Count of Events (Accidents + Incidents)")

ax.set_title("Distribution of Event by Flight Purpose in the USA, May 1, 2018 through April 30, 2023");
```

Finally, let's see distribution of event across aircraft type as well as events classified as accidents vs incidents over the last 5 years in the USA.

```python
# set up plots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 5))

# create variables for easier reuse
last_5_USA_aircraft_type_counts = df_unique_aircraft_events_usa_last_5['aircraft category'].value_counts()
last_5_USA_event_type_counts = df_unique_aircraft_events_usa_last_5['ev_type'].value_counts()

# plot data
ax1.bar(last_5_USA_aircraft_type_counts.index, last_5_USA_aircraft_type_counts.values)
ax2.bar(last_5_USA_event_type_counts.index, last_5_USA_event_type_counts.values)

# customize appearance
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=45)
ax1.set_ylabel("Count of Events (Accidents + Incidents)")
ax2.set_ylabel("Count of Events")
ax1.set_title("Distribution of Event by Aircraft Type, May 1, 2018 through April 30, 2023(USA only)")
ax2.set_title("Accidents vs Incidents, May 1, 2018 through April 30, 2023 (USA only)");
```
