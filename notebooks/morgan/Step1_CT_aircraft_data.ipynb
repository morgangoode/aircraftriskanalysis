The weather conditions seemed like something that would be a contributing factor for an accident. As we were still considering location-based recommendations, we wanted to use weather conditions to help inform that decision. But first, we needed to understand what these abbreviations stood for. 

Definitions are as follows: 


* IMC : Instrument Meteorological Conditions: these require the pilot to fly primarily by reference to instruments. Pilots using instruments to guide their flight will also take guidance and instructions from air traffic control.
* UNK : Unknown
* VMC : Visual Meteorological Conditions govern whether it is safe for pilots flying by sight. The pilot is only allowed to fly in specific weather expectations when not using instruments as guides. 



````python
# creating unique list of weather conditions 
weather_condition_list = df_fly_ct['Weather.Condition'].tolist() 
unique_weather_condition_list = set(weather_condition_list)
unique_weather_condition_list
````
